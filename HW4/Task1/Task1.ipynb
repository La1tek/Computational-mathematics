## Построим график функции $xe^x+x^2+1$ чтобы локализовать корни

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x*np.exp(x) + x**2 - 1

xs = np.linspace(-3, 2, 10000)
ys = f(xs)

plt.figure(figsize=(8,5))
plt.plot(xs, ys, label=r'f(x) = $x e^x + x^2 - 1$')
plt.axhline(0, color='black', linewidth=1)

plt.vlines([-1.2, -1.15, 0.45, 0.5], ymin=min(ys), ymax=max(ys), color='red', linestyles='--')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('График f(x)')
plt.grid(True)
plt.legend()

plt.show()
```

Получаем две области локкализации: 
- $x_1 \in [-1.2, -1.15]$
- $x_2 \in [0.45, 0.5]$

Для каждого метода простой итерации проверим достаточные условия:
Метод $x=g(x)$ сходится на $[a,b]$ если:
- $g([a,b]) \subseteq [a,b]$
- $|g'(x)|<1 \quad \forall x \in [a,b]$

```python
# Вспомогательная функция
# Принимает функию и отрезок
def plot_g(func, segment):
    a = segment[0]
    b = segment[1]
    xs = np.linspace(a, b, 10000)
    ys = func(xs)
    
    plt.figure(figsize=(8,5))
    plt.plot(xs, ys)
    
    plt.hlines([a, b], xmin=min(xs), xmax=max(xs), color='red', linestyles='--')
    
    plt.xlabel('x')
    plt.ylabel('g(x)')
    plt.title('График g(x)')
    plt.grid(True)
    plt.show()
```

### a) $g(x)=ln(1/x−x)$

```python
g = lambda x: np.log(1 / x - x)
plot_g(g, [-1.2, -1.15])
plot_g(g, [0.45, 0.5])
```

$g([-1.2,-1.15]) \not\subseteq [-1.2,-1.15] \Rightarrow$ о сходимости МПИ к $x_1$ говорить ничего нельзя

$g([0.45,0.5]) \not\subseteq [0.45,0.5] \Rightarrow$ о сходимости МПИ к $x_2$ говорить ничего нельзя

### б) $g(x)=\sqrt{1-xe^x}$

```python
g = lambda x: np.sqrt(1 - x * np.exp(x))
plot_g(g, [-1.2, -1.15])
plot_g(g, [0.45, 0.5])
```

$g([-1.2,-1.15]) \not\subseteq [-1.2,-1.15] \Rightarrow$ о сходимости МПИ к $x_1$ говорить ничего нельзя

$g([0.45,0.5]) \not\subseteq [0.45,0.5] \Rightarrow$ о сходимости МПИ к $x_2$ говорить ничего нельзя

### в) $g(x)=(1-x^2)e^{-x}$

```python
g = lambda x: (1 - x ** 2) * np.exp(-x)
plot_g(g, [-1.2, -1.15])
plot_g(g, [0.45, 0.5])
```

$g([-1.2,-1.15]) \not\subseteq [-1.2,-1.15] \Rightarrow$ о сходимости МПИ к $x_1$ говорить ничего нельзя

$g([0.45,0.5]) \not\subseteq [0.45,0.5] \Rightarrow$ о сходимости МПИ к $x_2$ говорить ничего нельзя

### г) $g(x)=1/x-e^x$

```python
g = lambda x: 1 / x - np.exp(x)
plot_g(g, [-1.17, -1.16])
plot_g(g, [0.45, 0.5])
```
