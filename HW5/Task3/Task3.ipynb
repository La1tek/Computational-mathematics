```python
import numpy as np

x = np.array([0.5, 0.6, 0.8, 1.0])
f = np.array([-0.378, -0.225, 0.103, 0.460])

def lagrange(x, f):
    n = len(x)
    L = np.zeros(n)  # коэффициенты l_j(0)

    for j in range(n):
        lj = 1.0
        for m in range(n):
            if m != j:
                lj *= (-f[m]) / (f[j] - f[m])
        L[j] = lj

    X0 = np.sum(x * L)
    return X0, L

root, coeffs = lagrange(x, f)

for i, c in enumerate(coeffs):
    print(f"l_{i}(0) = {c:.12f}")

print(f"x ≈ {root:.12f}")
```
